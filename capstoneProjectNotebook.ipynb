##  Next, I need to get the data from Wikipedia and tidy it up as required for the submission. I've left comments explaining what steps I took.

```python
# Second, we need the data from wikipedia. I've used the easiest route in pandas!

# Read the table on wikipedia 
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')[0]

# Removed Boroughs that are 'Not assigned'
df2 = df.loc[~(df['Borough'] == 'Not assigned')]

# Aggregated Neighbourhoods that share a Postal Code.
df2 = df2.groupby('Postal Code').agg({'Borough':'first',
                                   'Neighborhood': ', '.join,}).reset_index()
df2
```

```python
# The number of rows in the dataframe.
df2.shape[0]
```

### Submission Part 2 - Adding Coordinates

```python
# I've used the .csv file for this part
coordinatesDF = pd.read_csv('Geospatial_Coordinates.csv')

# I've used a Left Join here to add the coordinated to the previous dataframe.
merged_df = pd.merge(left=df2, right=coordinatesDF, how='left', left_on='Postal Code', right_on='Postal Code')
merged_df
```

### Submission Part 3 - Explore and Cluster the Neighbourhoods in Toronto

```python
#First, picking out Borough with 'Toronto' in the name.

df3 = merged_df[merged_df['Borough'].str.contains('Toronto', regex=False)]

# Now visualising all the data using Folium

map_toronto = folium.Map(location=[43.651070,-79.347015], zoom_start=10)

for lat,lng,borough,neighborhood in zip(df3['Latitude'],df3['Longitude'],df3['Borough'],df3['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
map_toronto
```

### Now clustering the neighbourhoods using KMeans clustering

```python
k=5
t_clustering = df3.drop(['Postal Code', 'Borough', 'Neighborhood'], 1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(t_clustering)
kmeans.labels_
df3.insert(0, 'Cluster Label', kmeans.labels_)
```

```python
df3
```

### Finally, representing the clusters on a map

```python
# create map
cluster_map = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for latitude, longitude, neighborhood, cluster in zip(df3['Latitude'], df3['Longitude'], df3['Neighborhood'], df3['Cluster Label']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [latitude, longitude],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(cluster_map)
# show map       
cluster_map
```
